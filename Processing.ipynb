```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.functions import col, from_json
```

```python

try:
    spark.sql("create catalog streamingdata;")
except Exception as e:
    print('check if catalog already exists',e)

try:
    spark.sql("create schema streamingdata.bronze;")
except:
    print('check if bronze schema already exists')

try:
    spark.sql("create schema streamingdata.silver")
except:
    print('check if silver schema already exists')

try:
    spark.sql("create schema streamingdata.gold;")
except:
    print('check if gold schema already exists')
```

Fair data

```python
connectionString = ""
eventHubName = "first"
consumergroup="first-consumer"

ehConf = {
  'eventhubs.connectionString': sc._jvm.org.apache.spark.eventhubs.EventHubsUtils.encrypt(connectionString),
  'eventhubs.eventHubName': eventHubName,
  'eventhubs.consumerGroup': consumergroup,
}

```

```python
df = spark.readStream \
    .format("eventhubs") \
    .options(**ehConf) \
    .load() \


# Displaying stream: Show the incoming streaming data for visualization and debugging purposes
df.display()

# Writing stream: Persist the streaming data to a Delta table 'streaming.bronze.weather' in 'append' mode with checkpointing
df.writeStream\
    .option("checkpointLocation", "/mnt/streamingdata/bronze/taxifair")\
    .outputMode("append")\
    .format("delta")\
    .toTable("streamingdata.bronze.taxifair")
```

```python
json_schema = StructType([
    StructField("medallion", StringType()),
    StructField("hack_license", StringType()),
    StructField("vendor_id", StringType()),
    StructField("rate_code", IntegerType()),
    StructField("store_and_fwd_flag", StringType()),
    StructField("pickup_datetime", StringType()),
    StructField("dropoff_datetime", StringType()),
    StructField("passenger_count", IntegerType()),
    StructField("trip_time_in_secs", IntegerType()),
    StructField("trip_distance", DoubleType()),
    StructField("pickup_longitude", DoubleType()),
    StructField("pickup_latitude", DoubleType()),
    StructField("dropoff_longitude", DoubleType()),
    StructField("dropoff_latitude", DoubleType())
])
```

```python
df = spark.readStream\
    .format("delta")\
    .table("streamingdata.bronze.taxifair")\
    .withColumn("body", col("body").cast("string"))\
    .withColumn("body",from_json(col("body"), json_schema))\
    .select("body.medallion", "body.hack_license", "body.vendor_id", "body.rate_code", "body.store_and_fwd_flag", "body.pickup_datetime","body.dropoff_datetime","body.passenger_count","body.trip_time_in_secs","body.trip_distance","body.pickup_longitude","body.pickup_latitude","body.dropoff_longitude","body.dropoff_latitude",col("enqueuedTime").alias('timestamp'))


# Displaying stream: Visualize the transformed data in the DataFrame for verification and analysis
df.display()

# # Writing stream: Save the transformed data to the 'streaming.silver.weather' Delta table in 'append' mode with checkpointing for data reliability
df.writeStream\
    .option("checkpointLocation", "/mnt/streaming_Data/silver/taxifair")\
    .outputMode("append")\
    .format("delta")\
    .toTable("streamingdata.silver.taxifair")
```

Trip Data

```python
connectionString = ""
eventHubName = "second"
consumergroup="second-consumer"

eh = {
  'eventhubs.connectionString': sc._jvm.org.apache.spark.eventhubs.EventHubsUtils.encrypt(connectionString),
  'eventhubs.eventHubName': eventHubName,
  'eventhubs.consumerGroup': consumergroup,
}
```

```python
df1 = spark.readStream \
    .format("eventhubs") \
    .options(**eh) \
    .load() \


# Displaying stream: Show the incoming streaming data for visualization and debugging purposes
df1.display()

# # Writing stream: Persist the streaming data to a Delta table 'streaming.bronze.weather' in 'append' mode with checkpointing
df1.writeStream\
    .option("checkpointLocation", "/mnt/streaming_Data/gold/taxi")\
    .outputMode("append")\
    .format("delta")\
    .toTable("streamingdata.bronze.taxitrip")
```

```python
json_trip_schema = StructType([
    StructField("medallion", StringType()),
    StructField("hack_license", StringType()),
    StructField("vendor_id", StringType()),
    StructField("pickup_datetime", StringType()),
    StructField("payment_type", StringType()),
    StructField("fare_amount", DoubleType()),
    StructField("surcharge", DoubleType()),
    StructField("mta_tax", DoubleType()),
    StructField("tip_amount", DoubleType()),
    StructField("tolls_amount", DoubleType()),
    StructField("total_amount", DoubleType())
])
```

```python
hold = spark.readStream\
    .format("delta")\
    .table("streamingdata.bronze.taxitrip")\
    .withColumn("body", col("body").cast("string"))\
    .withColumn("body",from_json(col("body"), json_trip_schema))\
    .select("body.medallion", "body.hack_license", "body.vendor_id","body.pickup_datetime","body.payment_type","body.fare_amount","body.surcharge","body.mta_tax","body.tip_amount","body.tolls_amount","body.total_amount",col("enqueuedTime").alias('timestamp'))

# Displaying stream: Visualize the transformed data in the DataFrame for verification and analysis
hold.display()

# Writing stream: Save the transformed data to the 'streaming.silver.weather' Delta table in 'append' mode with checkpointing for data reliability
hold.writeStream\
    .option("checkpointLocation","/mnt/streaming_Data/silver/taxi")\
    .outputMode("append")\
    .format("delta")\
    .toTable("streamingdata.silver.taxitrip")
```

Gold Layer

```python
#JOIN TABLE

fare = spark.readStream.table("streamingdata.silver.taxifair").withWatermark("timestamp", "5 minutes")
trip = spark.read.table("streamingdata.silver.taxitrip").withWatermark("timestamp", "5 minutes")

fare_selected = fare.select("medallion","rate_code","store_and_fwd_flag","dropoff_datetime","passenger_count","trip_time_in_secs","trip_distance","pickup_longitude","pickup_latitude","dropoff_longitude","dropoff_latitude","timestamp")

trip_selected = trip.select("medallion","hack_license","vendor_id","pickup_datetime","payment_type","fare_amount","surcharge","mta_tax","tip_amount","tolls_amount","total_amount")

query = (fare_selected
  .join(trip_selected, ["medallion"], "inner")
  .writeStream
  .option("checkpointLocation", "/mnt/streaming/gold_data/taxifare_summarys")
  .format("delta")
  .toTable("streamingdata.silver.combined")
)
```

```python
# AVERAGE AND TOTAL TRIPS
df2 = spark.readStream\
    .format("delta")\
    .table("streamingdata.silver.combined")\
    .withWatermark("timestamp", "5 minutes") 

df2.createOrReplaceTempView("streaming_average")

agg_query = """
    SELECT 
        avg(trip_distance) as Trip_Distance,
        avg(fare_amount) as Fare_Amount,
        count(*) as Total_trip
    FROM 
        streaming_average
    GROUP BY window(timestamp, "5 minutes")
"""

# Create a DataFrame from the SQL query
agg_df = spark.sql(agg_query)


# Writing Aggregated Stream: Store the aggregated data in 'streaming.gold.weather_aggregated' with checkpointing for data integrity
agg_df.writeStream\
    .option("checkpointLocation", "/mnt/streaming/golds_data/average_taxi")\
    .outputMode("complete")\
    .format("delta")\
    .toTable("streamingdata.gold.average")
```

```python
# Payment TYPE DONE
streaming_df = spark.readStream \
    .format("delta")\
    .table("streamingdata.silver.combined")\
    .withWatermark("timestamp", "5 minutes")
    

# Create a temporary view for the streaming DataFrame
streaming_df.createOrReplaceTempView("streaming_data")



agg_query = """
    SELECT 
        payment_type,
        count(payment_type) as num
    FROM 
        streaming_data
    group by payment_type
"""

# Create a DataFrame from the SQL query
agg_df = spark.sql(agg_query)

# Write the result DataFrame to the table in Delta format
agg_df.writeStream \
    .outputMode("complete")\
    .option("checkpointLocation", "/mnt/streaming/gold_data/taxi_payment_checkpoints") \
    .format("delta") \
    .toTable("streamingdata.gold.taxi_payment_type") 
```

```python
# Busy Day DONE
streaming_busyDay = spark.readStream \
    .format("delta")\
    .table("streamingdata.silver.combined")\
    .withWatermark("timestamp", "5 minutes")
    

# Create a temporary view for the streaming DataFrame
streaming_busyDay.createOrReplaceTempView("streaming_busyDay")



agg_querys = """
    SELECT
        COUNT(*) AS total,
        CASE DAYOFWEEK(dropoff_datetime)
            WHEN 1 THEN 'Sunday'
            WHEN 2 THEN 'Monday'
            WHEN 3 THEN 'Tuesday'
            WHEN 4 THEN 'Wednesday'
            WHEN 5 THEN 'Thursday'
            WHEN 6 THEN 'Friday'
            WHEN 7 THEN 'Saturday'
        END AS day_of_week
    FROM
        streaming_busyDay
    GROUP BY
        day_of_week;
"""

# Create a DataFrame from the SQL query
agg_df = spark.sql(agg_querys)

# Write the result DataFrame to the table in Delta format
agg_df.writeStream \
    .outputMode("complete")\
    .option("checkpointLocation", "/mnt/streaming/gold_data/taxi_buysDay_checkpoints") \
    .format("delta") \
    .toTable("streamingdata.gold.taxi_BusyDay") 
```

```python
# BUSSIEST MONTH DONE

streaming_busyMonth = spark.readStream \
    .format("delta")\
    .table("streamingdata.silver.combined")\
    .withWatermark("timestamp", "5 minutes")
    

# Create a temporary view for the streaming DataFrame
streaming_busyMonth.createOrReplaceTempView("streaming_busyMonth")

agg_querys = """
    SELECT  
       count(*) As total,
       month(dropoff_datetime) as Months
    FROM 
        streaming_busyMonth
    group by Months
"""

# Create a DataFrame from the SQL query
agg_df = spark.sql(agg_querys)

# Write the result DataFrame to the table in Delta format
agg_df.writeStream \
    .outputMode("complete")\
    .option("checkpointLocation", "/mnt/streaming/gold_data/taxi_buysMonth_checkpoint") \
    .format("delta") \
    .toTable("streamingdata.gold.taxi_BusyMonth") 
```

```python
#Count trips DONE
streaming_CountTrip = spark.readStream \
    .format("delta")\
    .table("streamingdata.silver.taxifair")\
    .withWatermark("timestamp", "5 minutes")
    

# Create a temporary view for the streaming DataFrame
streaming_CountTrip.createOrReplaceTempView("streaming_CountTrip")



agg_querys = """
    SELECT 
        COUNT(*) AS Total_trip,
        month(dropoff_datetime) as Months
    FROM 
        streaming_CountTrip
    group by Months
"""

# Create a DataFrame from the SQL query
agg_df = spark.sql(agg_querys)

# Write the result DataFrame to the table in Delta format
agg_df.writeStream \
    .outputMode("complete")\
    .option("checkpointLocation", "/mnt/streaming/gold_data/taxi_countTrips_checkpoint") \
    .format("delta") \
    .toTable("streamingdata.gold.taxi_CountTrips") 
```
